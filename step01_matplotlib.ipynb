**plot와 scatter의 차이점
 - scatter는 각 점의 속성(크기,색상등) 개별적으로 

```python
rng = np.random.RandomState(0) # 이 설정으로 인해 test 데이터를 동일하게 처리 가능
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

#alpha=0.3  겹치는 효과를 위한 투명도
#cmap='viridis' : 생략시 흑색
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar();  # 색상 척도 표시
```

## 파이차트 
 1. 카테고리별 값의 상

```python
labels = ['A','B','C','D']
sizes = [15,30,50,50]
colors = ['yellow','blue','red','lightcoral']


explode = (0, 0.1, 0, 0)
fig1, ax1 = plt.subplots()

ax1.pie(sizes,explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True, startangle=90)

ax1.axis('equal')
plt.show()
```
