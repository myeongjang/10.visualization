```python
data = np.random.randn(1000)
data
```

```python
plt.hist(data)
```

hist()
1.bins - 구간의 경계
- alpha - 불투명도(1), 투명도(0)
- edgecolor - 간선 표현을 위한 설정

```python
plt.hist(data, bins=100, alpha=0.3)
```

```python
plt.hist(data, bins=100, alpha=0.3, edgecolor="black")
```

```python
plt.hist(data,bins=5, alpha=0.3)
```

```python
plt.hist(data, bins=5, alpha=0.3, edgecolor='black')
plt.hist(data, bins=50, alpha=0.3, edgecolor='red')
```
