```python
data = np.random.randn(1000)
data
```

```python
plt.hist(data)
```

hist()
1. bins - 구간의 경계
- alpha - 불투명도(1), 투명도(0)
- edgecolor - 간선 표현을 위한 설정

```python
plt.hist(data, bins=100, alpha=0.3)
```

```python
plt.hist(data, bins=100, alpha=0.3, edgecolor='black')
```

```python
plt.hist(data, bins=5, alpha=0.3)
```

```python
plt.hist(data, bins=5, alpha=0.3, edgecolor='black')
plt.hist(data, bins=50, alpha=0.3, edgecolor='red')
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
```

# seaborn 학습
- seaborn 자체적으로 제공하는 dataset 활용해보기

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
iris = sns.load_dataset('iris')
iris
```

### 속성별로 어떤 연관성이 있는지 확인 가능한 그래프

```python
# hue='species'  ->  품종 즉 카테고리 컬럼명
sns.pairplot(iris, hue='species', markers=['o','s', 'D'], size=3)
```
