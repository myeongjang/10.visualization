```python
# 0-5까지 1씩 증가하는 변수
# x, y 축으로 사용 단 색은 초록색
xy =np.arange(1,5,1)
plt.plot(xy,xy,'gp')
```

```python
plt.style.use('seaborn')
plt.plot(xy,xy,marker="o",color='g', ls="dashed", linewidth=5)
```

```python
plt.plot(xy,xy,marker="o",color='g', ls="--", lw=5,ms=12)
```

```python
#x = np.arange(1,5,1)
x = np.linspace(0,10,100)
plt.plot(x,x+1,ls = 'solid' )
plt.plot(x,x+2,ls = 'dashed')
plt.plot(x,x+3,ls = 'dashdot')
plt.plot(x,x+4,ls = 'dotted')
```

```python
plt.plot(x,x+1,ls = '-' )
plt.plot(x,x+2,ls = '-.')
plt.plot(x,x+3,ls = '--')
plt.plot(x,x+4,ls = ':')
```

플롯에 타이틀과 레이블 추가

```python
plt.plot(x,x,ls = ':')
plt.plot(x,x+1,ls = '-' )
plt.plot(x,x+2,ls = '-.')
plt.plot(x,x+3,ls = '--')

plt.title('data view')

plt.xlim(0, 5)
plt.ylim(1, 5)
plt.show()
```

```python
plt.plot([1,2,3], [1,4,9], 'ro', label = 'line2', ls = '-')
# 주의사항 : 여러개의 플롯을 동시 작업시 범례(라벨표기)
# 구분을 위한 필수 함수
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3], [1,4,9], 'bs', label = 'line2', ls = '-')
plt.plot([4,5,6], [2,6,8], 'ro', label = 'line1', ls = '-.')
plt.legend()
plt.show()
```

```python
plt.plot([1,2,3], [1,4,9], 'bs', label = 'line2', ls = '-')
plt.plot([4,5,6], [2,6,8], 'ro', label = 'line1', ls = '-.')
# 축상의 위치 표시 지점 의미
plt.xticks([0,5,10])
plt.yticks([0,5,15])

plt.legend()
plt.show()
```

```python
#
data = np.random.RandomState(0)
data.rand(5)
```

```python
'''?
1.제시된 마커 표기로 list 생성
2.반복문으로 list내에 있는 마커를 포함
'''
markers = ['o','.',',','x','+','v','^','<','>','s','d']

rng = np.random.RandomState(0)
for i in markers:
    plt.plot(rng.rand(5),rng.rand(5),i, label= '"marker.{0}"'.format(i))
plt.legend()
plt.show
```
